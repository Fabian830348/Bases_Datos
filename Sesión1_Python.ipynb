# <FONT SIZE=5 COLOR="purple"> 1. Tipos de Datos en Python </FONT>

En python tenemos los siguientes tipos de datos

- **Integer**: Tipo de dato para números enteros, ya sean positivos o negativos.

- **Float**: Tipo de dato para números con precisión decimal, pueden ser negativos o positivos.

- **String**: Tipo de dato para cadena de texto con caracteres alfanuméricos. En Python se colocan entre comillas.

## <FONT SIZE=5 COLOR="green"> 5.1 Listas en Python </FONT>

- Las listas son objetos que permiten almacenar más de un dato.

- Almacenan información de una sola variable

- Para generar una lista usamos : [ ] o *list([ ])*

Veamos algunos ejemplos

```python
lista1 = [4,6,7,1,2,3,4,10]
print(lista1)
```

```python
lista2 = list([3,6,7,1,2,5,9,0,18])
print(lista2)
```

A continuación, algunas funciones útiles en listas

- ***len(lista)***: tamaño de la lista, es decir número de entradas

- ***type(lista)***: tipo de elementos de la lista.

- ***lista.append(elemento)***: agrega un elemento al final de la lista

- ***lista.extend(lista)***: añade una lista al final de otra lista.

- ***lista.insert(posición, elemento)***: inserta un elemento en una determinada posición de la lista.

- ***lista.reverse()***: invierte el orden de los elementos de la lista.

- ***lista.sort()***: ordena los elementos de la lista.

- ***lista.pop(n):*** elimina el elemento de la posición $n$.

Veamos algunos ejemplos

```python
# longitud de una lista
print(len([3,5,15,10,8]))
```

```python
#agregar un elemento al final
u=[3,5,15,10,8]
u.append("fabian")
u
```

```python
# agregar un elemento al final
u=[3,5,15,10,8]
u.append(["fabian", "Hugo"])
print(u)
print(len(u))
```

```python
# agregar una lista al final
u=[3,5,15,10,8]
u.extend(["fabian", "Hugo"])
print(u)
print(len(u))
```

```python
# invierte el orden
u=[3,5,15,10,8]
u.reverse()
u
```

```python
# ordena de menor a mayor
u=[3,5,15,10,8]
u.sort()
u
```

```python
# ordena de menor a mayor
u=[3,5,15,10,8]
u.sort(reverse= True)
u
```

```python
# ordena por orden alfabético
u=["fabian", "hugo" , "luis", "andres", "Luis" , "Andres"]
u.sort()
u
```

```python
# ordena de mayor a menor.
u=["fabian", "hugo" , "luis", "andres", "Luis" , "Andres"]
u.sort(reverse = True)
u
```

```python
# elimina el elemento de la posición 3
u=[3,5,15,10,8]
u.pop(3)
u
```

Otras expresiones en listas:

- $lista1 + lista2$ : concatena las listas

- $lista1*n$ : replicar la lista $n$ veces

- $min(lista)$: extrae el valor mínimo de la lista

- $max(lista)$: extrae el valor máximo de la lista

- $count(lista)$: cuenta las veces que aparece un elemento en la lista.

- $sum(lista)$: suma los elementos de la lista

Veamos algunos ejemplos

```python
# concatenar
print([3,4,5]+[3,4,6])
# replicar lista 3 veces
print([3,4,5]*3)
# calcula el mínimo
print(min([4,9,10,-6]))
# calcula el máximo
print(max([4,9,10,-6]))
# calcula la suma
print(sum([4,9,10,-6]))
# cuenta cuántas veces aparece 10
print([4,9,10,10,7,10,-6].count(10))
```

## <FONT SIZE=5 COLOR="purple"> 5.2 Diccionarios en Python </FONT>

- Es una estructura de datos para trabajar conjuntos de datos que están almacenados en pares del tipo

$$clave : valor$$

- Para acceder a los datos del diccionario se debe hacer por medio de la clave, que debe ser única.

- Una característica de los diccionarios es su representación en forma de llaves $\{$  $\}$

$$dicc = \{``clave_1" : datos_1 , ``clave_2" : datos_2 \dots ``clave_n" : datos_n \}$$

También existen otras formas de representación

$$dicc = dict([(``clave_1", datos_1) , (``clave_2", datos_2) \dots (``clave_1", datos_n)])$$

Veamos algunos ejemplos



```python
# se puede crear usando {} así:
dicc = {"Nombre" : "Fabian",
         "Edad" : 24,
         "Sexo" : "Masculino",
         "Cédula" : 80987656}
print(dicc)
```

```python
# se puede crear usando la instrucción dict([])
dicc1 = dict([("Nombre", "fabian"),
           ("Edad", 24),
           ("Sexo", "Masculino"),
           ("Cédula", 80097865)])
print(dicc1)
```

```python
# se puede crear usando dict() así:
dicc2 = dict(Nombre = "Fabian",
          Edad = 24,
          Sexo = "Masculino",
          Cédula = 80987656)
print(dicc2)
```

Las claves no necesariamente deben tener un solo elemento

```python
dicc3 = dict(Nombre = ["fabian", "andres", "carlos"],
          Edad = [23,34,45])
print(dicc3)
```

**Acceder y modificar los datos del diccionario**

- Para acceder a un elemento del diccionario usamos la instrucción:

$$dicc[``clave"]$$

Veamos algunos ejemplos.

```python
# {'Nombre': 'Fabian', 'Edad': 24, 'Sexo': 'Masculino', 'Cédula': 80987656}
# accedemos a la clave "nombre"
print(dicc["Nombre"])
```

```python
# {'Nombre': ['fabian', 'andres', 'carlos'], 'Edad': [23, 34, 45]}
# accedemos a la clave "nombre"
print(dicc3["Nombre"])
```

```python
# ¿Qué hace la siguiente estructura?
for elemento in dicc2:
  print(elemento)
```

```python
# ¿Qué hace la siguiente estructura?
for elemento in dicc2:
  print(dicc2[elemento])
```
