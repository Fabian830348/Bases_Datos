# <FONT SIZE=5 COLOR="green"> 1. Definición de DataFrame en Python </FONT>

- En este notebook trabajaremos uno de los objetos más importante cuando manipulamos y procesamos datos, los DataFrames.

- Recordemos que existen dos grupos grandes de datos que se denominan:

  - **Datos estructurados**: son aquellos que se pueden representar en un arreglo bidimensional: *fila-columna*. Es decir, las clásicas hojas de cálculo de excel.
  - **Datos no estructurados**: son los que no podemos representar en forma estructurada *fila-columna*. Por ejemplo, archivos de vídeo, audio, texto, imágenes, etc.


 De acuerdo con lo anterior definimos:

 ***Dataframe***: son arreglos o estructuras bidimensionales compuestos por fila y por columnas, y cuyos datos pueden ser de diferentes tipos: números, cadenas de texto, booleanos, etc.

   - Un dataframe tiene forma de matriz, sin embargo, el primero admite datos de diferentes tipos y en segundo solamente de un tipo.
   - Un dataframe está compuesto por vectores (columnas) cada una de las cuales describe una atributo o característica (variable) de una población.
   - La librería más usada para manipulación de dataframes se denomina ***pandas*** en Python. Si trabajas en *RStudio* es ***dplyr***

# <FONT SIZE=5 COLOR="green"> 3. Instrucciones básicas</FONT>

Alguna instrucciones básicas de una dataframe:

**df.info()** : Devuelve información (número de filas, número de columnas, índices, tipo de las columnas y memoria usado) sobre el DataFrame df.

**df.shape** : Devuelve una tupla con el número de filas y columnas del DataFrame df.

**df.size** : Devuelve el número de elementos del DataFrame.

**df.columns** : Devuelve una lista con los nombres de las columnas del DataFrame df.

**df.index** : Devuelve una lista con los nombres de las filas del DataFrame df.

**df.dtypes** : Devuelve una serie con los tipos de datos de las columnas del DataFrame df.

**df.head(n)** : Devuelve las n primeras filas del DataFrame df.

**df.tail(n)** : Devuelve las n últimas filas del DataFrame df.

**df.append({})** : para agregar una nueva fila al dataframe

Para eliminar una columna o una fila usamos la instrucción .drop

$$df.drop(``nombre", axis = 1, inplace = T \, or \, F)$$

- *axis = 0* : es por  fila

- *inplace = F* : no sobreescribe. Mantiene el original

$$df.set\_index(columnas, drop = True, inplace = False)$$

# <FONT SIZE=5 COLOR="green"> 4. Acceder a los elementos del DataFrame </FONT>

Para acceder a los elementos, o a las filas o columnas del DataFrame podemos usar:

- ***iloc*** : para acceder por la posición
- ***loc.*** : para acceder por el nombre

Si usamos la primera instrucción

- ***df.iloc[i,j]*** : elemento de la fila i y columna j.

- ***df.iloc[i]*** : devuelve la fila i

- ***df.iloc[lista i , lista j]*** : devuelve un data.frame con las filas en lista i y columnas en columna j.

- ***df.iloc[:, j]*** : devuelve toda columna j.

Si usamos la instrucción **.loc**

- ***df.loc[nombre fila , nombre columna]***: es lo mismo anterior pero con los nombres.

- ***df[nombre columna]*** : devuelve las columnas en name columnas

- ***df.columna*** : otra forma para acceder a una columna que en el nombre no tenga espacios en blanco.

Para los ejemplos que sigue construiremos el siguiente DataFrame a partir de unas listas dadas.

```python
lista1 = ["Fabian" , 30 , 164 , "Doctorado", 1300, "soltero"]
lista2 = ["Camilo" , 24 , 174 , "Maestría",2300,"casado"]
lista3 = ["Ana María" , 20 , 180 , "Profesional",4000, "casado"]
lista4 = ["Andrea" , 36 , 154 , "Doctorado",5000,"soltero"]
lista5 = ["Mariana" , 45 , 164 , "Doctorado",6000, "casado"]
lista6 = ["Tatiana" , 40 , 184 , "Doctorado",3200, "casado"]
lista7 = ["Stiven" , 22 , 168 , "Profesional",2000, "soltero"]
lista8 = ["Edwin" , 35 , 167 , "Maestría",5500, "soltero"]
lista9 = ["Karem" , 28 , 162 , "Maestría",4800, "casado"]
lista10 = ["Gustavo" , 50 , 180 , "Profesional",3500, "soltero"]
columns = ["Nombre", "Edad", "Estatura", "Nivel_Educativo", "Salario", "Estado Civil"]
datos = pd.DataFrame([], columns=columns)
A=[globals()[f"lista{i}"] for i in range(1,11)]
for i in range(10):
  dic=dict(zip(columns,A[i]))
  datos = datos.append(dic, ignore_index=True)
print(datos)
```

```python
datos[["Edad", "Estatura" , "Salario"]] = datos[["Edad", "Estatura" , "Salario"]].astype(float)
```

```python
datos.info()
```

## <FONT SIZE=4 COLOR="magenta"> 5.1 Ordenar según una variable </FONT>

Para ordenar por una variable usamos la instrucción


$$df.sort\_values(variable, ascending = False) $$

se usa:

- *ascending=True* :  ordena de menor a mayor.
- *ascending=False* : ordena de mayor a menor.


```python
# ordenamos por la variable Edad de menor a mayor
datos.sort_values("Edad", ascending = True)
```

```python
# ordenamos por la variable Estatura de mayor a menor.
datos.sort_values("Estatura", ascending = False)
```

```python
# ordenamos por la variable Nombre en orden alfabético
datos.sort_values("Nombre", ascending = True)
```

## <FONT SIZE=4 COLOR="magenta"> 5.2 Renombrar una variable</FONT>

Podemos renombrar una columna de la siguiente forma usando la instrucción ***rename***

$$ df.rename(columns={key : value, key : value}, inplace=True) $$

 - Para cambiar el nombre de una variable debemos usar un diccionario donde las claves son las variables actuales y los valores son los nombres nuevos de las variables.
 - Recordemos que la instrucción **inplace** se usa para determinar si cambiamos el DataFrame original.




```python
# cambiaremos el nombre de Nivel_Educativo por Escolaridad
datos.rename(columns={"Nivel_Educativo" : "Escolaridad"}, inplace=True)
datos
# cuidado: a partir de esta línea de código debemos trabajar con la nueva variable renombrada.
```

## <FONT SIZE=4 COLOR="magenta"> 5.5 Agrupaciones: ***groupby()*** </FONT>

- Se usa para agrupar una variable por sus categorías y luego aplicar una función a estos grupos con respecto a otra variable

<center><img src="https://github.com/Fabian830348/cursos/blob/master/intro/groupby1.png?raw=true" alt="centered image" width="450" height="350"></center> <center><figcaption> <FONT SIZE=1 COLOR="black"> Fuente: Elaboración propia  </FONT> <figcaption></center>




```python
datos.groupby("Escolaridad")
```

Observe que no se genera nada. Esto es porque internamente lo que hace pandas es agrupar por las clases o categorias de la variable.

Podemos acceder a los grupos con la instrucción

$$datos.groupby(``Variable").get\_group(``categoría")$$

```python
# agrupa por las categorias de la variable Escolaridad y selecciona el grupo Doctorado
datos.groupby("Escolaridad").get_group("Doctorado")
```

Podemos que se haga una división por grupos de dos variables

```python
# hacemos agrupación por dos variables deben colocarse en forma de lista
# firts() es para ver las primeras categorias de la variable inicial
datos.groupby(["Escolaridad", "Estado Civil"]).first()
```

Podemos hacer operaciones con las agrupaciones, así:

colocar imagen

$$datos.groupby(``Variable")["segunda variable].funcion()$$

Por ejemplo, supongamos que queremos agrupar por escolaridad y calcular el promedio de las edades de cada uno de los grupos.

<center><img src="https://github.com/Fabian830348/cursos/blob/master/intro/groupby2.png?raw=true" alt="centered image" width="450" height="350"></center> <center><figcaption> <FONT SIZE=1 COLOR="black"> Fuente: Elaboración propia  </FONT> <figcaption></center>

```python
# se agrupa por escolaridad y en cada grupo se calcula la media de la edad.
datos.groupby("Escolaridad")["Edad"].mean()
```

```python
# agrupa por escolaridad, toma el máximo de las estaturas y los ordena de mayor a menor.
datos.groupby("Escolaridad")["Estatura"].max().sort_values(ascending=False)
```

```python
datos.groupby(["Escolaridad", "Estado Civil"])["Estatura"].count()
```

# <FONT SIZE=5 COLOR="green"> 6. Algunos filtros con DataFrame </FONT>

En esta sección revisaremos algunos filtros en un dataframe. Para esto, generaremos algunas preguntas y buscaremos responderlas con los filtros.

```python
# personas con más de 38 años
datos[datos.Edad>38]
```

```python
# si queremos solo el conteo
datos[datos.Edad>38].count()[0]
```

```python
# personas que miden mayor o igual a 160 y tienen doctorado
print(datos[(datos.Estatura>=170) & (datos.Escolaridad == "Doctorado")])
print(datos[(datos.Estatura>=170) & (datos.Escolaridad == "Doctorado")].count()[0])
```

```python
# personas solteras con maestría y más de 30 años
datos[(datos["Estado Civil"]=="soltero") & (datos.Escolaridad=="Maestría") & (datos.Edad > 30)]
```

Cuando no existe una función genérica para aplicar directamente, podemos aplicar una de la siguiente manera.

$$datos.groupby(``variable1")[``variable2"].apply(lambda \quad x: function)$$


```python
# personas cuyo nombre empieza por C y ganan más de 4.5 millones
datos[datos.Nombre.apply(lambda x: x.startswith("A")) & (datos.Salario>4500)]
```

```python
datos.groupby("Escolaridad")["Estatura"].apply(lambda x: (x-x.mean())/np.std(x))
```
